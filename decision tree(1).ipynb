```python
#filter the values which contains more than 24.99 relative humidity at 3pm

clean_data=data.copy()
clean_data['high_humidity_label']=(clean_data['Humidity3pm']>24.99)*1
clean_data['high_humidity_label'].head()
```

```python
clean_data=data.copy()
clean_data['high_humidity_label']=(clean_data['Humidity3pm']>24.99)*1
clean_data['high_humidity_label'].tail()
```

```python
y=clean_data[['high_humidity_label']].copy()
y.head()
```

```python
clean_data['Humidity3pm'].head()
```

```python
y.head()
```

```python
#use 9am sensor signals at features to predict humidity at 3pm

morning_features=['MinTemp' ,'MaxTemp', 'Rainfall','Evaporation', 'Sunshine', 'WindGustSpeed','WindSpeed9am', 'WindSpeed3pm', 'Humidity9am', 'Humidity3pm', 'Pressure9am', 'Pressure3pm', 'Cloud9am', 'Cloud3pm', 'Temp9am', 'Temp3pm', 'RISK_MM']
```

```python
x=clean_data[morning_features].copy()
x.columns
```

```python
y.columns
```

```python
#perform test and train split


x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.33,random_state=365)
```

```python
humidity_classifier=DecisionTreeClassifier(max_leaf_nodes=10,random_state=0)
humidity_classifier.fit(x_train,y_train)
```

```python
type(humidity_classifier)
```

```python
#predict on test set

y_predicted=humidity_classifier.predict(x_test)
```

```python
y_predicted[:10]
```

```python
y_test['high_humidity_label'][:10]
```

```python
#measure accuracy of the classifier

accuracy_score(y_test,y_predicted)*100
```
